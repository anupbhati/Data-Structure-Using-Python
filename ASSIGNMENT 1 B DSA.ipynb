```python
# Calculator
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b): return a / b if b != 0 else "Error"
def power(a, b): return a ** b

a = 5
b = 3

print("Addition:", add(a, b))
print("Subtraction:", subtract(a, b))
print("Multiplication:", multiply(a, b))
print("Division:", divide(a, b))
print("Power:", power(a, b))


# Discount Calculator
def calculate_discount(price, discount):
    return price - (price * discount / 100)

price = 1000
discount = 20
print("Discounted Price:", calculate_discount(price, discount))


# Grade Evaluator
def evaluate_grade(marks):
    if marks >= 90: return "A"
    elif marks >= 75: return "B"
    elif marks >= 60: return "C"
    elif marks >= 50: return "D"
    else: return "F"

marks = 85
print("Grade:", evaluate_grade(marks))


# Even/Odd Checker
def check_even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

num = 7
print("Even/Odd:", check_even_odd(num))


# Array Operations
arr = [1, 2, 3, 4, 5]

# Swap
arr[1], arr[3] = arr[3], arr[1]
print("After Swapping:", arr)

# Split
print("Split:", arr[:2], arr[2:])

# Insert
arr.insert(2, 99)
print("After Inserting:", arr)

# Delete
arr.remove(3)
print("After Deleting:", arr)
```
